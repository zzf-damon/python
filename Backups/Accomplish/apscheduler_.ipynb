# 简单的定时任务

```python
# fixme 间隔性任务
# -*- coding: utf-8 -*-
import os
from apscheduler.schedulers.blocking import BlockingScheduler

def test():
    from time import sleep
    print("开始程序")
    sleep(5)
    print("结束程序")

def test1():
    from time import sleep
    print("开始程序1")
    sleep(2)
    print("结束程序1")

def test2():
    from time import sleep
    print("开始程序2")
    sleep(1)
    print("结束程序2")

def run_timing():
    job_defaults = {'max_instances': 3}
    scheduler = BlockingScheduler(timezone='MST', job_defaults=job_defaults)
    # scheduler.add_job(main, 'interval', hours=3)
    # scheduler.add_job(get_unrelated, 'interval', hours=1)
    scheduler.add_job(test, 'interval', seconds=2)
    scheduler.add_job(test1, 'interval', seconds=1)
    scheduler.add_job(test2, 'interval', seconds=1)
    print(
        'Press Ctrl+{0} to exit'.format('Break' if os.name == 'nt' else 'C    '))

    try:
        scheduler.start()
    except (KeyboardInterrupt, SystemExit):
        pass


if __name__ == '__main__':
    run_timing()
    # p("a")
```

# 不影响主任务的定时任务

```python
from apscheduler.schedulers.background import BackgroundScheduler

import time


def job():
    print('job 3s')

    time.sleep(5)

    print("5s")



def run():
    job_defaults = {'max_instances': 2}

    sched = BackgroundScheduler(timezone='MST', job_defaults=job_defaults)

    sched.add_job(job, 'interval', id='3_second_job', seconds=3)

    # sched.start()
    try:
        sched.start()
    except (KeyboardInterrupt, SystemExit):
        pass

if __name__ == '__main__':
    run()


    while (True):
        print('main 1s')

        time.sleep(1)
```
